```python
raw_data = open('day1data.txt', 'r').readlines()
```

```python
numbers = []
for line in raw_data:
    s = line.strip()
    if s == "":
        n = 0
    else:    
        n = int(s)
    numbers.append(n)
```

```python
numbers[0:5]
```

```python
elf_sums = []
current_sum = 0
for n in numbers:
    
    if n == 0:
        elf_sums.append(current_sum)
        current_sum = 0
    else:
        current_sum = current_sum + n

elf_sums.append(current_sum)
```

```python
elf_sums
```

```python
max(elf_sums)
```

```python
elf_sums.sort(reverse = True)
```
