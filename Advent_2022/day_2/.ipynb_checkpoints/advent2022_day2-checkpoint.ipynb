```python
def play_ver2(opp, player):
    
    win = 6
    draw = 3
    loss = 0

    if opp == player:
        return draw + player
    elif opp == player-1 or opp == player +2:
        return win + player
    elif opp == player+1 or opp == player-2:
        return loss + player
```

```python
def play_by_guide(opp, outcome):
    
    win = 6
    draw = 3
    loss = 0
    
    rock = 1
    paper = 2
    scissors = 3
    
    if outcome == 2:
        return opp + draw
    
    elif outcome ==1 and opp == rock:
        return scissors + loss
    elif outcome ==1 and opp == paper:
        return rock + loss
    elif outcome ==1 and opp == scissors:
        return paper + loss
    
    elif outcome == 3 and opp == rock:
        return paper + win
    elif outcome == 3 and opp == paper:
        return scissors + win
    elif outcome == 3 and opp == scissors:
        return rock + win
```
